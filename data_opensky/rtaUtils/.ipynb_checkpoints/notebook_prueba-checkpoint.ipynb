## Nearest neighbour sorting algorithm test

```python

data = np.array([[0,20,1,2,6,16,5,3, 8, 15],[1,6,4,9,7,0,2,3,8,10]]).transpose()

# print(data)

def distance2(lat1: np.array, lon1: np.array, lat2: np.array,
              lon2: np.array) -> np.array:
    dlat = lat2 - lat1
    dlon = lon2 - lon1

    return np.sqrt(np.power(dlat, 2) + np.power(dlon,2))

def presort(array: np.array):
    distances = []
    for x in array:

        distances.append(haversine_np(array[:,0],  array[:,1], x[0], x[1]))
        # print(distance2(array[:,0],  array[:,1], x[0], x[1]))

    # print(array)

    distances = np.array(distances)
    mins = [0]
    mask = np.zeros(shape=array.shape[0]).astype(bool)
    while True:
        mask[mins[-1]] = True

        x = np.ma.array(distances[mins[-1]], mask=mask)
        mins.append(x.argmin(fill_value=10000))
        if len(mins) == array.shape[0]:
            break
        
    # print(mins)

    # print(array[mins])

    return array[mins]
```

```python

sns.scatterplot(pd.DataFrame(data),x=0, y=1, ax=sns.lineplot(pd.DataFrame(data),x=0, y=1, sort=False))
```

```python

# presort(data)

sns.scatterplot(pd.DataFrame(presort(data)), x=0, y=1, ax=sns.lineplot(pd.DataFrame(presort(data)),x=0, y=1, sort=False))
```

```python
found, missing = set(),set()
inicial, final = np.array([np.nan]*pepe.shape[0]*2),  np.array([np.nan]*pepe.shape[0]*2)

for row, (i, j) in pepe.iterrows():

    if i in missing:
        missing.remove(i)

    if j in found:
        found.remove(j)
        
    if (i+len(found)) == (j+len(missing)):
        final[j+len(missing)-1] = j
        inicial[i+len(found)-1] = i
        
    elif row+len(missing)+1 == i:
        missing.add(j)
        inicial[i+len(found)-1] = i
        final[j+len(missing)-1] = j
        
        print(row+1,'missing')
        
    elif row+len(found)+1 == j:
        found.add(i)
        inicial[i+len(found)-1] = i
        final[j+len(missing)-1] = j
        
        print(row+1,'found')
    
    else:
        print('patata')
    

    
    
    print(found, missing)
print(inicial)
print(final)
```
